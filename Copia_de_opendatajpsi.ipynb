```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm


pc = pd.read_csv('https://goo.gl/g8y9TC')
pc['Masa_Invariante'] = ((pc.E1+pc.E2)**2 - ((pc.px1+pc.px2)**2 + (pc.py1+pc.py2)**2 + (pc.pz1+pc.pz2)**2))**.5
pc.head()

len(pc)


jpsy = pc[(pc.Masa_Invariante >= 2.95) & (pc.Masa_Invariante <= 3.2)]
xx = pc[(pc.Masa_Invariante >= 3.2) & (pc.Masa_Invariante <= 3.6)]
xc0 = pc[(pc.Masa_Invariante >= 3.6) & (pc.Masa_Invariante <= 3.75)]
xy = pc[(pc.Masa_Invariante >= 3.75) & (pc.Masa_Invariante <= 3.9)]
fig = plt.figure(figsize=(15, 10))

plt.hist(jpsy.Masa_Invariante, 400, range = (2.9,3.9), facecolor='r')
plt.hist(xc0.Masa_Invariante, 400, range = (2.9,3.9), facecolor='g')
plt.hist(xx.Masa_Invariante, 400, range = (2.9,3.9), facecolor='b')
plt.hist(xy.Masa_Invariante, 400, range = (2.9,3.9), facecolor='b')
plt.xlabel('Invariant mass (GeV/$c^2$)', fontsize = 15)
plt.ylabel('Number of events \n', fontsize = 15)
plt.title('Measured distribution of events with two muons \n', fontsize = 15)

plt.show()


fig = plt.figure(figsize=(16, 12))
plt.hist(pc.eta1, 500, range = (-5,5), facecolor='g')
plt.xlabel('Pseudorapidity (GeV/$c^2$)', fontsize = 15)
plt.ylabel('Number of events \n', fontsize = 15)
plt.title('Measured distribution of events with one muon \n', fontsize = 15)


plt.show()
```
